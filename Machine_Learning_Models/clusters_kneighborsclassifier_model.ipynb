```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/Project 4 MN Home-loan Predictive Model (HMDA Data-source)/\
Clean_Data_For_Model/Cleaned_loan_data_for_model.csv')

df.head()
```

```python
# Seperate the features, X,  from the target variable, y
y = df['action_taken']
X = df.drop(columns='action_taken')
```

```python
# Preview the features data
X.head()
```

```python
# Preview the first five entries for the target variable
y[:5]
```

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
```

```python
# Review the features data
X.head()
```

```python
# Split the dataset using train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Instantiate a StandardScaler instance
scaler = StandardScaler()

# Fit the training data to the standard scaler
X_scaler = scaler.fit(X_train)

# Transform the training data using the scaler
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Import the KNeighborsClassifier module from sklearn
from sklearn.neighbors import KNeighborsClassifier

# Instantiate the KNeighborsClassifier model with n_neighbors = 3
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model using the training data
knn.fit(X_train_scaled, y_train)
```

```python
# Create predictions using the testing data
y_pred = knn.predict(X_test_scaled)
```

```python
# Print the classification report comparing the testing data to the model predictions
print(classification_report(y_test, y_pred))
```

Key Observations:
- Strong performance on the majority class (approvals)
- Room for improvement in identifying denials
- Significant class imbalance (128 denials vs 19,439 approvals)
- Model may benefit from techniques to address class imbalance
